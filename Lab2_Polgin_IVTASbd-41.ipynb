## Постановка задачи

Разработать систему для анализа параметров **скорости** и **расстояния** с использованием функций принадлежности и вычисления пересечений нечетких множеств.  

### Необходимо реализовать следующие этапы:
1. Определить функции принадлежности для параметров «Скорость» и «Расстояние» с помощью треугольных функций.  
2. Вычислить степени принадлежности для заданных значений скорости и расстояния.  
3. Рассчитать пересечения между функциями принадлежности.  
4. Построить графики функций принадлежности и их пересечений.



## Описание предметной области

Система предназначена для анализа **уровня риска** в зависимости от скорости движения объекта и расстояния до препятствия.

**Входные параметры:**
- **Скорость (км/ч):** «Медленно», «Умеренно», «Быстро», «Опасно быстро».  
- **Расстояние (м):** «Близко», «Средне», «Далеко».  

Цель системы — вычислить пересечение этих параметров для оценки уровня риска.



## Функции принадлежности

### Для скорости:
1. «Медленно»: треугольная функция [0, 40], пик при 20  
2. «Умеренно»: треугольная функция [30, 70], пик при 50  
3. «Быстро»: треугольная функция [60, 100], пик при 80  
4. «Опасно быстро»: треугольная функция [90, 2³¹−1], пик при 110  

### Для расстояния:
1. «Близко»: треугольная функция [0, 10], пик при 5  
2. «Средне»: треугольная функция [8, 20], пик при 15  
3. «Далеко»: треугольная функция [18, 2³¹−1], пик при 30  



## Программная реализация

Программа реализована на языке **Python** с использованием библиотек **NumPy** и **Matplotlib**.


```

import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')

speed_min, speed_max = 0, 2**31 - 1
distance_min, distance_max = 0, 2**31 - 1

def triangular_membership(x, a1, a, a2):
    if x < a1 or x > a2:
        return 0
    elif a1 <= x < a:
        return (x - a1) / (a - a1)
    elif a <= x <= a2:
        return (a2 - x) / (a2 - a)
```

```

def intersect_memberships(speed_values, distance_values):
    intersections = {}
    for s_name, s_val in speed_values.items():
        for d_name, d_val in distance_values.items():
            intersections[(s_name, d_name)] = min(s_val, d_val)
    return intersections
```

```

def plot_memberships():
    x_vals_speed = np.linspace(0, 130, 200)
    x_vals_distance = np.linspace(0, 40, 200)

    plt.figure(figsize=(10, 6))
    plt.plot(x_vals_speed, [triangular_membership(x, 0, 20, 40) for x in x_vals_speed], label="Медленно")
    plt.plot(x_vals_speed, [triangular_membership(x, 30, 50, 70) for x in x_vals_speed], label="Умеренно")
    plt.plot(x_vals_speed, [triangular_membership(x, 60, 80, 100) for x in x_vals_speed], label="Быстро")
    plt.plot(x_vals_speed, [triangular_membership(x, 90, 110, speed_max) for x in x_vals_speed], label="Опасно быстро")
    plt.title("Функции принадлежности скорости")
    plt.xlabel("Скорость (км/ч)")
    plt.ylabel("Значение принадлежности")
    plt.legend()
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.plot(x_vals_distance, [triangular_membership(x, 0, 5, 10) for x in x_vals_distance], label="Близко")
    plt.plot(x_vals_distance, [triangular_membership(x, 8, 15, 20) for x in x_vals_distance], label="Средне")
    plt.plot(x_vals_distance, [triangular_membership(x, 18, 30, distance_max) for x in x_vals_distance], label="Далеко")
    plt.title("Функции принадлежности расстояния")
    plt.xlabel("Расстояние (м)")
    plt.ylabel("Значение принадлежности")
    plt.legend()
    plt.show()
```

```

def plot_intersections_3d(intersections, speed_labels, distance_labels):
    matrix = np.zeros((len(speed_labels), len(distance_labels)))
    for (s, d), val in intersections.items():
        matrix[speed_labels.index(s), distance_labels.index(d)] = val

    X, Y = np.meshgrid(range(len(distance_labels)), range(len(speed_labels)))

    fig = plt.figure(figsize=(9, 6))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, matrix, cmap='coolwarm', edgecolor='k')
    ax.set_title("3D график пересечений функций принадлежности")
    ax.set_xlabel("Расстояние")
    ax.set_ylabel("Скорость")
    ax.set_zlabel("Пересечение")
    ax.set_xticks(range(len(distance_labels)))
    ax.set_yticks(range(len(speed_labels)))
    ax.set_xticklabels(distance_labels)
    ax.set_yticklabels(speed_labels)
    fig.colorbar(surf, shrink=0.5)
    plt.show()
```

```

def main():
    speed_values = {
        "Медленно": triangular_membership(35, 0, 20, 40),
        "Умеренно": triangular_membership(35, 30, 50, 70),
        "Быстро": triangular_membership(70, 60, 80, 100),
        "Опасно быстро": triangular_membership(100, 90, 110, speed_max)
    }

    distance_values = {
        "Близко": triangular_membership(8, 0, 5, 10),
        "Средне": triangular_membership(15, 8, 15, 20),
        "Далеко": triangular_membership(19, 18, 30, distance_max)
    }

    intersections = intersect_memberships(speed_values, distance_values)

    print("Результаты пересечения:")
    for (s, d), val in intersections.items():
        print(f"{s:12} × {d:8} → {val:.2f}")

    plot_memberships()
    plot_intersections_3d(intersections,
                          ["Медленно", "Умеренно", "Быстро", "Опасно быстро"],
                          ["Близко", "Средне", "Далеко"])

main()
```
